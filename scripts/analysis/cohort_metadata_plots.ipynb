```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from hne.core.paths import PREPROCESSED_COHORT

df = pd.read_csv(PREPROCESSED_COHORT / "metadata.csv")

df.head()
```

```python
sns.set()
sns.displot(df['mean_tumor_fraction'], bins=30, kde=True, aspect=2)
```

```python
mean_tumor_tiles = df['n_tumor_tiles_saved'].mean()
sns.set()
p = sns.displot(df["n_tumor_tiles_saved"], bins=30, kde= True, aspect=2)

p.ax.axvline(x=mean_tumor_tiles, color='red', linestyle='--', label=f'Mean: {mean_tumor_tiles:.1f}')

p.ax.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 8))

sns.scatterplot(
    data=df,
    x='n_tumor_tiles',
    y='n_initial_tiles'
)

plt.title("Initial tiles vs final tumor tiles per patient")
plt.xlabel("Tumor Tiles")
plt.ylabel("Initial Tiles")
plt.show()
```
